```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Generation of Plots for VGG

```python
vgg_df = pd.read_csv('../wandb_results/vgg_cifar10_wandb_export.csv')
```

```python
len(['pt_test_acc']*400)
```

```python
len(list(vgg_df['accuracy_diff']))
```

```python
all_stats = list(vgg_df['pt_test_acc'])
all_stats.extend(list(vgg_df['tf_test_acc']))
all_stats.extend(list(vgg_df['accuracy_diff']))

all_names = (['PT Test Accuracy']*400)
all_names.extend(['TF Test Accuracy']*400)
all_names.extend(['Accuracy Diff']*400)
```

```python
vgg_stats_df = pd.DataFrame({'Value': all_stats, 'Statistic': all_names})
```

```python
plt.rcdefaults()
```

```python
import matplotlib.pyplot as plt

SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 22

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE, figsize=(1, 1))  # fontsize of the figure title
```

```python
viridis = sns.color_palette("viridis")
viridis_colors = [viridis[0], viridis[2], viridis[5]]
```

```python
# sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
max(vgg_stats_df['Value'])
```

```python
ax = sns.displot(vgg_stats_df, x="Value", hue="Statistic", kind="kde", cut=0, bw_adjust=1, palette=viridis_colors, linewidth=2.5, legend=False, aspect=1.5)
# plt.setp(ax._legend.get_texts(), fontsize='14') # for legend text
# plt.setp(ax._legend.get_title(), fontsize='16')
# ax._legend._loc_real = 1
plt.xlim(0, 0.85)
plt.ylim(0, 2.8)
plt.axvline(0.79610002, c="crimson", lw=2.5)
plt.tight_layout()
ax.savefig('vgg_stat_plot_with_max_acc.png', dpi=200, bbox_inches="tight")
```

### Generation of DenseNet Plots

```python
densenet_df = pd.read_csv('../wandb_results/densenet_cifar10_wandb_export.csv')
```

```python
dense_all_stats = list(densenet_df['pt_test_acc'])
dense_all_stats.extend(list(densenet_df['tf_test_acc']))
dense_all_stats.extend(list(densenet_df['accuracy_diff']))

dense_all_names = (['PT Test Accuracy']*400)
dense_all_names.extend(['TF Test Accuracy']*400)
dense_all_names.extend(['Accuracy Diff']*400)
```

```python
dense_stats_df = pd.DataFrame({'Value': dense_all_stats, 'Metric': dense_all_names})
```

```python
max(dense_stats_df['Value'])
```

```python
# plt.rcParams['legend.loc'] = 'upper right'
ax = sns.displot(dense_stats_df, x="Value", hue="Metric", kind="kde", cut=0, bw_adjust=1, palette=viridis_colors, linewidth=2.5, aspect=1)
plt.setp(ax._legend.get_texts(), fontsize='14') # for legend text
plt.setp(ax._legend.get_title(), fontsize='16')
plt.xlim(0, 0.85)
plt.ylim(0, 2.8)
plt.axvline(0.7759999979734421, c="crimson", lw=2.5)
ax._legend._loc_real = 9
plt.tight_layout()
ax.savefig('dense_stat_plot_with_max_acc.png', dpi=200)
```
